```python
import pandas as pd
import numpy as np
```

```python
customers = pd.read_csv("Customer_Feedback_Data.csv")
products = pd.read_csv("Product_Offering_Data.csv")
transactions = pd.read_csv("Transaction_Data.csv")
```

```python
transactions['Transaction_Amount'] = transactions['Transaction_Amount'].fillna(transactions['Transaction_Amount'].median())
```

```python
transactions.isnull().sum()
```

```python
customers['Feedback_Comments'] = customers['Feedback_Comments'].fillna("Unknown")
```

```python
customers.isnull().sum()
```

```python
customers['Satisfaction_Score'] = customers['Satisfaction_Score'].fillna(customers['Satisfaction_Score'].mean())
customers['Satisfaction_Score'] = customers['Satisfaction_Score'].fillna(customers['Satisfaction_Score'].median())
customers.dropna(subset=['Satisfaction_Score'], inplace=True)
```

```python
customers.isnull().sum()
```

```python
customers.drop_duplicates(inplace=True)
transactions.drop_duplicates(inplace=True)
products.drop_duplicates(inplace=True)
```

```python
# Customers
print("Duplicates in customers:", customers.duplicated().sum())
customers.drop_duplicates(inplace=True)

# Transactions
print("Duplicates in transactions:", transactions.duplicated().sum())
transactions.drop_duplicates(inplace=True)

# Products
print("Duplicates in products:", products.duplicated().sum())
products.drop_duplicates(inplace=True)
```

```python
transactions['Transaction_Date'] = pd.to_datetime(transactions['Transaction_Date'])
```

```python
products['Product_Type'] = products['Product_Type'].str.title().str.strip()
products['Risk_Level'] = products['Risk_Level'].str.upper().str.strip()
```

```python
transactions = pd.DataFrame({
    "Transaction_ID": [1, 2, 3],
    "Customer_ID": [101, 102, 101],
    "Transaction_Date": ["2026-01-01", "2026-01-02", "2026-01-03"],
    "Transaction_Amount": [100, 200, 150],
    "Transaction_Type": ["Purchase", "Purchase", "Purchase"],
    "Product_ID": ["P001", "P002", "P001"]  # <-- Add this column
})
```

```python
df = customers.merge(transactions, on='Customer_ID', how='left')

df['Product_ID'] = df['Product_ID'].astype(str)
products['Product_ID'] = products['Product_ID'].astype(str)

df_full = df.merge(products, on="Product_ID", how="left")



```

```python
products.columns
```

```python
df.head()
```

```python
# Total transaction amount per customer
df_total = df.groupby('Customer_ID')['Transaction_Amount'].sum().reset_index()
df_total.rename(columns={'Transaction_Amount': 'TotalTransactionAmount'}, inplace=True)

# Number of transactions per customer
df_count = df.groupby('Customer_ID')['Transaction_ID'].count().reset_index()
df_count.rename(columns={'Transaction_ID': 'NumTransactions'}, inplace=True)

# Merge totals and counts
df_summary = df_total.merge(df_count, on='Customer_ID')
df_summary.head()
```

```python
df_feedback = df.groupby('Customer_ID')['Satisfaction_Score'].mean().reset_index()
df_feedback.rename(columns={'Satisfaction_Score':'AvgSatisfactionScore'}, inplace=True)

df_summary = df_summary.merge(df_feedback, on='Customer_ID', how='left')

df_summary.head()
```

```python
# Account age (days since first transaction)
first_tx = df.groupby('Customer_ID')['Transaction_Date'].min().reset_index()
first_tx.rename(columns={'Transaction_Date':'FirstTransactionDate'}, inplace=True)
df = df.merge(first_tx, on='Customer_ID', how='left')
df['AccountAgeDays'] = (pd.to_datetime('2026-01-27') - df['FirstTransactionDate']).dt.days
_
```
